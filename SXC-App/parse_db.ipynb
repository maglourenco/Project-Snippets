```python
moradas = [result['results'][i]['address'] for i in range(0, res_size)]
```

```python
repeated = []
        
while len(moradas) > 0:
    current_val = moradas[0]
    for i in range(1, len(moradas)):
        if moradas[i] == current_val:
            repeated.append(moradas[i])
            moradas.remove(moradas[i])
    moradas.remove(current_val)
```
